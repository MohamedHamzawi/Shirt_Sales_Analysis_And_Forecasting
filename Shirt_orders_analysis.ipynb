```python
import numpy as np
import pandas as pd
```

## Load_Dataset

```python
Customers=pd.read_csv('customers.csv')
Products=pd.read_csv('Products.csv')
Orders=pd.read_csv('ORDERS.csv')
Customers.head()
```

```python
Products.head()
```

```python
Orders.head()
```

```python
df=pd.merge(Orders,Products,on='ProductID')
df.head()
```

```python
df=pd.merge(df,Customers,on='CustomerID')
df.head()
```

```python
df.drop(['Field1'],axis=1,inplace=True)
df.head()
```

```python
df.info()
```

## data cleaning

```python
df['State'].unique()
```

```python
df['Name'].unique()
```

```python
df['full_date'] = pd.to_datetime(df['OrderDate'], format='%d/%m/%Y %H:%M')
df.head()
```

```python
df['year']=df['full_date'].dt.year
df['month']=df['full_date'].dt.month
df['day']=df['full_date'].dt.day
df.head()
```

```python
df.drop(['ProductID','OrderDate','Street','Zip','Phone'],axis=1,inplace=True)
df.head()
```

```python
df['Revenue2'] = np.round(np.where(df['Discount'] == 0,
                         df['UnitsOrdered'] * df['UnitPrice'],
                         df['UnitsOrdered'] * df['UnitPrice'] * (1-df['Discount'])))
df.tail()
```

```python
df.rename(columns={'Revenue2':'Revenue'},inplace=True)
df.head()
```

## data Explorarity

```python
# revenue per year
df.groupby('year')['Revenue'].sum()
```

```python
# bar chart plot to plot revenue per year

import matplotlib.pyplot as plt
revenue_per_year = df.groupby('year')['Revenue'].sum()
years = revenue_per_year.index
revenue = revenue_per_year.values

plt.bar(years, revenue)
plt.xlabel('Year')
plt.ylabel('Revenue')
plt.title('Revenue per Year')
plt.show()
```

```python
# bar chart to plot orders per year

import matplotlib.pyplot as plt
orders_per_year = df.groupby('year')['UnitsOrdered'].sum()
years = orders_per_year.index
orders = orders_per_year.values

plt.bar(years, orders)
plt.xlabel('Year')
plt.ylabel('Orders')
plt.title('Orders per Year')
plt.show()
```

```python
#  bie chart to plot order by gender

import matplotlib.pyplot as plt
# group the data by gender and count the number of orders
orders_by_gender = df.groupby('Gender')['UnitsOrdered'].sum()

# create a pie chart
plt.pie(orders_by_gender.values, labels=orders_by_gender.index, autopct='%1.1f%%')
plt.title('Orders by Gender')
plt.show()
```

```python
#  line chart to compare betweeen months

import matplotlib.pyplot as plt
# Group the data by month and sum the revenue
revenue_by_month = df.groupby('month')['Revenue'].sum()


# Create a line chart
plt.bar(revenue_by_month.index, revenue_by_month.values)
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.title('Revenue by Month')
plt.show()
```

```python
# prompt: chart to plot revenue per state

import matplotlib.pyplot as plt
# Group the data by state and sum the revenue
revenue_by_state = df.groupby('State')['Revenue'].sum().sort_values(ascending=False)

# Create a bar chart
plt.bar(revenue_by_state.index, revenue_by_state.values)
plt.xlabel('State')
plt.ylabel('Revenue')
plt.title('Revenue by State')
plt.xticks(rotation=90)
plt.show()
```

```python
#  chart to plot most 10 shirts orders

import matplotlib.pyplot as plt
# Group the data by product name and sum the units ordered
product_orders = df.groupby('Name')['UnitsOrdered'].sum().sort_values(ascending=False)

# Select the top 10 most ordered products
top_10_products = product_orders.head(10)

# Create a bar chart
plt.bar(top_10_products.index, top_10_products.values)
plt.xlabel('Name')
plt.ylabel('Units Ordered')
plt.title('Top 10 Most Ordered Products')
plt.xticks(rotation=90)
plt.show()
```

```python
# plot best 5 customers

import matplotlib.pyplot as plt
# Group the data by customer ID and sum the revenue
customer_revenue = df.groupby('CustomerID')['Revenue'].sum().sort_values(ascending=False)

# Select the top 5 customers with the highest revenue
top_5_customers = customer_revenue.head(5)

# Create a bar chart
plt.bar(top_5_customers.index, top_5_customers.values)
plt.xlabel('Customer ID')
plt.ylabel('Revenue')
plt.title('Top 5 Customers by Revenue')
plt.xticks(rotation=90)
plt.show()
```

```python
from scipy.stats.distributions import chi2
def LLR_test(mod_1,mod_2,DF=1):
    L1 = mod_1.fit().llf
    L2 = mod_2.fit().llf
    LR = (2*(L2-L1))
    p = chi2.sf(LR,DF).round(3)
    return p
```
